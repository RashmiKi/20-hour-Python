```python
import pandas as pd
df = pd. read_csv(r"D:\Python\20-hour-Python\Matplotlib\data_set.csv")
df.head()
```

```python
df["Date"] = pd.to_datetime(df["Date"])

df["Month"] = df["Date"].dt.month_name()
df
```

```python
monthly_sales = df.groupby("Month")["Amount"].sum()
monthly_sales
```

```python
monthly_order = ["January", "February", "March", "April", "May", "June", "July", "August", "Spetember", "October", "November", "December"]

monthly_sales = monthly_sales.reindex(monthly_order).dropna()
```

```python
import matplotlib.pyplot as plt

plt.plot(monthly_sales.index, monthly_sales.values, marker = "D", color ="blue", linestyle = "--", linewidth = "2", label = "Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Amount")
plt.show()
```

```python
plt.bar(monthly_sales.index, monthly_sales.values, color = "Blue")
plt.xlabel("Month")
plt.ylabel("Amount")
plt.title("Monthly Sales")

for i,val in enumerate(monthly_sales.values):
    plt.text(i,val+1000,str(val), ha="center")

plt.show()
```

```python
plt.hist(df["Amount"], bins = 5, color = "Blue", edgecolor = "Orange")
plt.xlabel("Amount")
plt.grid()
plt.title("Amount Frequency")
plt.ylabel("Frequency")
plt.show()
```

```python
person_spend = df.groupby("Customer")["Amount"].sum()

plt.pie(person_spend, autopct= "%1.1f%%", labels= person_spend.index, startangle= 90)
plt.title("Sales Done by Customer")

plt.show()
```
