# 18/11/2025 : Probability Skewness

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skewnorm
```

```python
sns.set(style="whitegrid")
```

```python
normal = np.random.normal(loc=0,scale=1,size=10000)
right = skewnorm.rvs(a=10,loc=0,scale=1,size=10000)
left = skewnorm.rvs(a=-10,loc=0,scale=1,size=10000)
```

```python
plt.figure(figsize=(12,8))
sns.kdeplot(normal,label="Normal Distribution", linewidth=2)
sns.kdeplot(right, label="Right Skewed (Positive)", linewidth=2)
sns.kdeplot(left, label = "Left-Skewed(Negative)", linewidth=2)
plt.title("Normal vs Skewed Distribution", fontsize = 16)
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```
