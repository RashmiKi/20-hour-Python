# Feature Engineering

```python
# One-Hot Encoding

import pandas as pd
data = {'Color':['Red', 'Blue', 'Green', 'Blue']}
df = pd.DataFrame(data)

df_en = pd.get_dummies(df,columns=['Color'], prefix='Color')

print(df_en)
```

```python
# Binning
import pandas as pd

data = {'Age':[23,45,18,34,67,50,21]}
df = pd.DataFrame(data)

bins = [0,20,40,60,100]
labels = ['0-20','21-40','41-60', '61+']

df['Age_Group'] = pd.cut(df['Age'],bins = bins, labels=labels, right=False)

print(df)
```

# Scaling

```python
import pandas as pd
import numpy as np

df = pd.read_csv(r"D:\Python\20-hour-Python\Practice Rashmi\SampleFile.csv")

df = df.select_dtypes(include=np.number)
df.head()
```

```python
max = np.max(np.abs(df),axis=0)
scaled_df = df/max
scaled_df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler =MinMaxScaler()
scaled_data = scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled_data, columns=df.columns)

scaled_df.head()
```
