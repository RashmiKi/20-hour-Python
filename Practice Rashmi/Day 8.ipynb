# Detecting outliers

```python
import numpy as np
import pandas as pd
```

```python
data = pd.Series([10, 12, 11, 13, 12, 400, 9, 11])
```

```python
# step 1: compute z-scores
mean = data.mean()
std = data.std()
z_scores = (data - mean) / std

print("Z-scores:")
print(z_scores)
```

```python
# step 2: detect outliers (absolute z-score > 3)
outliers = data[np.abs(z_scores) > 3]
print("\nOutliers:")
print(outliers)
```

```python
# step 3: apply z-score normalization (turn all values into their z-scores)
normalized_data = z_scores
print("\nZ-score normalized data:")
print(normalized_data)
```
