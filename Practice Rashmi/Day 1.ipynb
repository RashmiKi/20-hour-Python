```python
# sum of numbers

numbers = [5,10,15,20]
s = sum(numbers)
print("Sum of Numbers:", s)
```

```python
number = [12,4,56,17,8]
m = max(number)
print("Maximum:", m)
```

```python
num = [10,20,30,40,50,60]
print(num[1:4])
print(num[4:8])
print(num[1:8:2])
```

```python
n = [1,2,3,4,5,6]
e = 0
od = 0

for i in n:
    if i%2 == 0:
        e = e+1
    else:
        od = od+1
print("Even:", e)
print("Odd:", od)

```

```python
# Reverse a List

numb = [1,2,3,4,5]
print(numb[::-1])
```

# Day 2 Conditional, Functions and String Operations

```python
#BMI Calculator

def bmi_calculator(height,weight):
    bmi = (weight/(height*height))
    print(bmi)

h = float(input("Enter your height: "))
w = float(input("Enter your weight: "))

bmi_calculator(h,w)
```

```python
#TIP calculator

def tip_Calculator(tp,p,bill):
    total_tip = ((tp*bill)/p)
    print("total tip per person:", total_tip)

tip = float(input("Enter tip percentage: "))
person = int(input("Number of humans in table: "))
bill = float(input("Enter total bill of food: "))

tip_Calculator(tip,person,bill)
```
