# Simple Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
X = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)  # independent variable
y = np.array([2, 4, 5, 4, 5, 7])               
```

```python
plt.scatter(X,y)
```

```python
# Create model
m = linear_model.LinearRegression()

# Fit model
m.fit(X, y)
```

```python
m.intercept_     # intercept
```

```python
m.coef_[0]       # slope
```

```python
y_pred = m.predict(X)
```

```python
y_pred
```

```python
# Plot actual vs predicted
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, y_pred, color='red', label='Regression line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()
```

```python
m.predict([[1.5]])
```

```python
m.predict([[4.6]])
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
from sklearn import datasets
```

```python
diabetes = datasets.load_diabetes()
```

```python
df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
df['target'] = diabetes.target
```

```python
df.head()
```

```python
X = df[['bmi']]   # independent variable
y = df['target']  # dependent variable
```

```python
plt.scatter(X, y)
```

```python
m = linear_model.LinearRegression()
m.fit(X, y)
```

```python
m.intercept_
```

```python
m.coef_[0]
```

```python
y_pred = m.predict(X)
```

```python
# Plot actual vs predicted
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, y_pred, color='red', label='Regression line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()
```

```python
m.predict([[0.06]])
```

```python
m.predict([[0.13]])
```
