```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('matplotlib.practice.csv')
```

```python
# Line plot - Monthly sales trend
plt.figure(figsize=(8,5))
plt.plot(data['Month'], data['Product_A'], label='Product A', marker='o')
plt.plot(data['Month'], data['Product_B'], label='Product B', marker='s')
plt.plot(data['Month'], data['Product_C'], label='Product C', marker='^')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Units Sold')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Bar chart - Advertising spend
plt.figure(figsize=(8,5))
plt.bar(data['Month'], data['Advertising_Spend'], color='skyblue', edgecolor='black')
plt.title('Monthly Advertising Spend')
plt.xlabel('Month')
plt.ylabel('Spend (in $000s)')
plt.show()
```

```python
# Scatter plot - Relationship between advertising and total sales
data['Total_Sales'] = data[['Product_A', 'Product_B', 'Product_C']].sum(axis=1)
plt.figure(figsize=(6,5))
plt.scatter(data['Advertising_Spend'], data['Total_Sales'], color='coral')
plt.title('Advertising Spend vs Total Sales')
plt.xlabel('Advertising Spend ($000s)')
plt.ylabel('Total Sales (Units)')
plt.grid(True)
plt.show()
```

```python
# Histogram - Customer satisfaction distribution
plt.figure(figsize=(6,4))
plt.hist(data['Customer_Satisfaction'], bins=5, color='lightgreen', edgecolor='black')
plt.title('Distribution of Customer Satisfaction')
plt.xlabel('Satisfaction Score')
plt.ylabel('Frequency')
plt.show()
```
