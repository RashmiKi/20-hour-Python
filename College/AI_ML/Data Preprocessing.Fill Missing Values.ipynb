```python
null_var = df.isnull().sum()/df.shape[0]*100
null_var
```

```python
sns.heatmap(df2.isnull())
```

```python
df3_num = df2.select_dtypes(include = ['int64', 'float64'])
df3_num
```

```python
sns.heatmap(df3_num.isnull())
```

```python
df4 = df3_num.fillna(df3_num.mean())
df4.isnull().sum()
```

```python
df5 = df3_num.fillna(df3_num.median())
df5.isnull().sum()
```
