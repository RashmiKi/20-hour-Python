```python
dataset = pd.read_csv(r"D:\Python\20-hour-Python\Data Analyst Course\tested.csv")
dataset.head()
```

```python
mn = dataset["Age"].mean()
```

```python
sns.histplot(x='Age', data=dataset, bins=[i for i in range(0,81,10)])
plt.plot([mn for i in range(0,300)], [i for i in range(0,300)])
plt.show()
```

```python
np.median(dataset["Age"])
```

```python
dataset["Age"].fillna(dataset["Age"].mean(),inplace=True)
```

```python
md = np.median(dataset["Age"])
```

```python
sns.histplot(x='Age', data=dataset, bins=[i for i in range(0,81,10)])
plt.plot([mn for i in range(0,300)], [i for i in range(0,300)])
plt.plot([md for i in range(0,300)],[i for i in range(0,300)], c ="red")
plt.show()
```

```python
mo = dataset['Age'].mode()[0]
```
